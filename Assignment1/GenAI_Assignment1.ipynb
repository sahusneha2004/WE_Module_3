```python
def pairwise(n: int) -> list[tuple[str, str]]:
 """
 Generates pairs of adjacent digits from a number.

 Args:
     n: The integer to generate pairs from.

 Returns:
     A list of tuples, where each tuple contains two adjacent digits as strings.
 """
 digits = str(n)
 return list(zip(digits[:-1], digits[1:]))
def describe(prev: int, curr: int) -> str:
  """
  Describes the order between two digits.

  Args:
      prev: The previous digit.
      curr: The current digit.

  Returns:
      A character representing the order:
          'A': Ascending (prev < curr)
          'D': Descending (prev > curr)
          'N': Neither ascending nor descending (equal)
  """
  if prev < curr:
    return 'A'
  elif prev > curr:
    return 'D'
  else:
    return 'N'

def compress(pattern: str) -> str:
  """
  Compresses a pattern by removing consecutive duplicate characters.

  Args:
      pattern: The pattern to compress.

  Returns:
      The compressed pattern.
  """
  result = []
  prev = None
  for char in pattern:
    if char != prev:
      result.append(char)
    prev = char
  return ''.join(result)

def classify_number(number):
  """
  Classifies a number based on the order of its digits.

  Args:
      number: The integer to classify.

  Returns:
      A character representing the classification:
          'A': Ascending order (strict)
          'D': Descending order (strict)
          'P': Peak order (ascending then descending)
          'V': Valley order (descending then ascending)
          'N': None of the above
  """
  digits = str(number)
  pattern = ''.join(describe(int(d1), int(d2)) for d1, d2 in pairwise(number))
  compressed_pattern = compress(pattern)
  if compressed_pattern == 'A':
    return 'A'
  elif compressed_pattern == 'D':
    return 'D'
#   elif compressed_pattern == 'N':
#     return 'N'  # All digits are the same #THIS IS NOT APPROPRIATE
  elif compressed_pattern == "AD":
    return 'P'
  elif compressed_pattern == "DA":
    return 'V'
  else:
    # Handle cases with mixed patterns (e.g., "ADAD" and "N" containing values)
    return 'N'
# Example usage (unchanged)
number = 12345  # Ascending
print(classify_number(number))  # Output: A

number = 54321  # Descending
print(classify_number(number))  # Output: D

number = 12321  # Peak
print(classify_number(number))  # Output: P

number = 32145  # Valley
print(classify_number(number))  # Output: V

number = 111  # Same digits
print(classify_number(number))  # Output: N

number = 2134  # Mixed pattern (ascending then descending)
print(classify_number(number))  # Output: P # WE WANT V OUTPUT FOR THIS

number = 43212  # Mixed pattern (descending then ascending)
print(classify_number(number))  # Output: V

number = 1234124521
print(classify_number(number))  # Output: N
```
